To do this, Xiaokang used an elastic net model. Let's try to make sure we can replicate his results using our Python implementation first, as a benchmark. 

````python
from sklearn import linear_model, preprocessing, ensemble
import numpy as np 
````

````python
feature_names = list(df.columns[0:55])
target_names = 'kcat km kcatkm'.split()
````

````python
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(10, 6))
for n, target in enumerate(target_names):
    measured = 'measured_{}'.format(target)
    cols = [measured] + feature_names
    d = df[cols].dropna()
    scaler = preprocessing.MinMaxScaler()
    X = pandas.DataFrame(scaler.fit_transform(d.iloc[:, 1:]), index=d.index, columns=d.columns[1:])
    y = d[measured].ravel()
    
    preds = pandas.DataFrame(None, index=range(10), columns=range(len(y)))
    for i, row in preds.iterrows():
        lm = linear_model.ElasticNetCV(l1_ratio=[0.5, 0.7, 0.9], cv=10, max_iter=1e4)
        clf = ensemble.BaggingRegressor(base_estimator=lm, oob_score=True) 
        clf.fit(X, y)
        preds.iloc[i] = clf.oob_prediction_
        
    d['average_prediction'] = list(preds.mean(axis=0))
    d.plot(x=measured, y='average_prediction', ax=ax[0, n], kind='scatter', color='black')
    d.hist(measured, ax=ax[1, n])
````

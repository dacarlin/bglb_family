```python
from Bio import SeqIO 
import pandas 
from random import choices 

def fasta_to_df(fasta_file):
    data = []
    for record in SeqIO.parse(fasta_file, 'fasta'):
        data.append(pandas.Series(list(record.seq), name=record.id))
    df = pandas.DataFrame(data) 
    return df 
        
df = fasta_to_df('/Users/alex/Downloads/PF00232_full.txt') 
df.shape
```

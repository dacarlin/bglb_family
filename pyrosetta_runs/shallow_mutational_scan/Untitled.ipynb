```python
import pyrosetta
from rosetta.protocols import simple_moves, residue_selectors, enzdes
from rosetta.core import select, chemical
import numpy as np
import sys
import seaborn as sns
import pandas
from Bio.Data import IUPACData
import matplotlib.pyplot as plt 

pyrosetta.init('-beta -extra_res_fa example_input/pnpg.params -run:preserve_header')
```

```python
pose = pyrosetta.pose_from_file('example_input/DMS_input.pdb')
score = pyrosetta.create_score_function('beta_cst')

wild_type_energy = score( pose )
print( 'Wild type energy:', wild_type_energy )
```

```python
def create_vector(pose):
    total_residue = pose.total_residue()
    en = pose.energies()
    score_types = score.get_nonzero_weighted_scoretypes()
    for i in range(1, total_residue+1):
        for j in range(1, len(score_types)+1):
            yield en.residue_total_energies(i)[score_types[j]]
            
len(list(create_vector(pose)))
```

```python
def generate_data_set(total_residue=400):
    for i in range(total_residue or 10):
        copy_pose = pose.clone()
        for k, v in IUPACData.protein_letters_1to3.items():
            protocol = [simple_moves.MutateResidue(i+1, v.upper()),]
            for item in protocol:
                item.apply(copy_pose)
            try:
                score(copy_pose)
                #lore = '{}{}'.format(i, k)
                yield create_vector(copy_pose)
            except:
                print("except")
            
df = pandas.DataFrame(generate_data_set())
df
```
